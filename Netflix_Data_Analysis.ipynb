### 16. Change 'duration' to int for films

```python
df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'] = df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'].str.replace(' min', '')
#print('Replace min', df_netflix.head(5))

df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'] = df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'].fillna('0')
#print('Fill NaN', df_netflix.head(5))

df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'] = df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'].astype(int)
#print('To Int', df_netflix.head(5))
```

### 17. Change 'duration' to int for series

```python
df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'] = df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'].str.replace('[^\d]', '', regex=True)
#print('Replace min', df_netflix.head(5))

df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'] = df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'].fillna('0')
#print('Fill NaN', df_netflix.head(5))

df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'] = df_netflix.loc[df_netflix['type'] == 'TV Show', 'duration'].astype(int)
#print('To Int', df_netflix.head(5))
```

j. La rpartition des directors des uvres franaises

```python
print(f"There are {df_netflix['country'].isnull().sum()} rows with no values for country.")
print(f"There are {df_netflix['director'].isnull().sum()} rows with no values for director.")

df_netflix['country']= df_netflix['country'].replace({pd.NA: 'NS'})
df_netflix['director']= df_netflix['director'].replace({pd.NA: 'NS'})

#Explosion and selection of french director
df_netflix['director'] = df_netflix['director'].str.split(', ')
df_exploded = df_netflix.explode('director')
```

```python
#plotly
df_france = df_exploded[df_exploded['country']=='France']

french_directors = df_france['director']

director_counts = french_directors.value_counts()

fig = go.Figure(go.Bar(
    x=director_counts.index,
    y=director_counts.values
))

fig.update_layout(
    title='Distribution of Directors for French Works',
    xaxis=dict(title='Director'),
    yaxis=dict(title='Count'),
    bargap=0.1  # Adjust the gap between bars
)

fig.show()
```

```python
#test for explode method
test= df_netflix.sample(n=350, random_state=42)
test['director']= test['director'].replace({pd.NA: 'NS'})
test['director'] = test['director'].str.split(', ')
df_exploded = test.explode('director')
df_exploded[df_exploded['country'].str.contains('France', na=False)]

france = df_exploded[df_exploded['country'].str.contains('France', na=False)]

french_directors = france['director']
director_counts = french_directors.value_counts()

plt.figure(figsize=(10, 6))
plt.scatter(director_counts.index, director_counts.values, color='red', alpha=0.8)
plt.title('Distribution of Directors of French Works')
plt.xlabel('Director')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.grid(True)
plt.tight_layout()
plt.show()
```
