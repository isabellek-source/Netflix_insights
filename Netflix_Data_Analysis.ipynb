### 10. Missingno, chart of missing values

"Missingno" is a term commonly used in data analysis, particularly in the Python programming language and its associated libraries like Pandas and Matplotlib. It refers to a visualization tool used to identify missing or null values in a dataset.

Missingno provides several visualizations to help you understand the patterns of missing data within your dataset:

1. Matrix plot
2. Bar chart
3. Heatmap
4. Dendogram 

1. Matrix plot : The matrix plot is a graphical representation of the completeness of the dataset. It displays the presence or absence of data as colored blocks, where missing values are represented by blank (white) spaces. This plot allows you to quickly identify regions of missing data and patterns of missingness.

```python
msno.matrix(df_netflix, color=(0.3, 0.2, 0.5))

plt.show()
```

Fact : In this matrix plot, the missing data in the 'date_added', 'rating', and 'duration' columns are not visible. We will investigate further in the following charts.

2. Bar Chart: The bar chart shows the distribution of missing values across variables. It displays the count or percentage of missing values for each column in the dataset, allowing you to identify which variables have the most missing data.

```python
msno.bar(df_netflix, color="skyblue")
```

As we see in this msno.bar(), missing values are identifiable through the column totals at the top. 

Here, the missing values for the three previously identified columns are visible

4. Dendrogram: The dendrogram is another visualization provided by Missingno, which displays the hierarchical clustering of variables based on their correlation in terms of missing values. It helps identify groups of variables that have similar patterns of missingness, which can be useful for understanding the underlying structure of missing data in the dataset.

```python
msno.dendrogram(df_netflix)
```

### 16. Change 'duration' to int for films

```python
df_netflix[df_netflix['duration'].isnull()]

# There are 3 values NaN for duration and it's from type=Movie. 
# We see that the duration are from the rating column. 
# This need to be corrected. 
```

```python
#### Correction ####
# Creating a filter to access to the rows with the values = ('74 min', '84 min', '66 min')
filtered_rows = df_netflix[(df_netflix['rating'] == '74 min') | (df_netflix['rating'] == '84 min') | (df_netflix['rating'] == '66 min')]

# Extraction of the duration values corresponding to the filtered rows
# Stores in a temporary variable temp_duration. 
# It uses .loc to access the 'duration' column for the filtered rows.
temp_duration = df_netflix.loc[filtered_rows.index, 'duration'].copy()

# Swaping the values of 'duration' and 'rating
df_netflix.loc[filtered_rows.index, 'duration'] = df_netflix.loc[filtered_rows.index, 'rating']

# Assigning the duration values stored in the temporary variable temp_duration 
# back to the 'rating' column for the filtered rows, effectively swapping the values back
df_netflix.loc[filtered_rows.index, 'rating'] = temp_duration

# Replacing any occurrences of 0 in the 'rating' column with None.
df_netflix['rating'] = df_netflix['rating'].replace(0, None)
```

```python
# Replacing str caracters from movies to '' (nothing), leaving just the number.
df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'] = df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'].str.replace(' min', '')
print('Replace min', df_netflix.head(5))
```

```python
# Convertion of the selected values to int, with .loc - accessor to select rows and columns based on the condition df_netflix['type'] == 'Movie', 'duration' 
df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'] = df_netflix.loc[df_netflix['type'] == 'Movie', 'duration'].astype(int)
print('To Int', df_netflix.head(5))

# Disclaimer : as long the column contains string values, the column isn't correctly converted to int. 
# To be continued ... 
```

### 19. Unique values for : 'type', 'country', 'release_year', 'rating', 'listed_in'

1. .value_counts(): 

This method returns a Series containing counts of unique values in descending order. 
It calculates the frequency of each unique value in the Series and returns the result as a new Series, 
where the index contains the unique values and the values contain their corresponding frequencies.

```python
print(df_netflix['type'].value_counts())
```

2. .unique(): 

This method returns an array of unique values in the Series. 
It essentially filters out duplicates and returns only the unique values in the Series. 
The values are returned in the order in which they appear in the Series.

```python
# Create a copy of the DataFrame to work with
df_netflix_copy = df_netflix.copy()

# Split every country with "comma"
df_netflix_copy['country'] = df_netflix_copy['country'].str.split(',')

# Explode the lists into multiple rows
df_netflix_copy = df_netflix_copy.explode('country')

# Remove spaces before and after the countries
df_netflix_copy['country'] = df_netflix_copy['country'].str.strip()

# Get the unique values from the 'country' column
unique_values_country = df_netflix_copy['country'].unique()

# Print the unique values
print(unique_values_country)
```

```python
## Listed_in
# Explode the lists into multiple rows
df_netflix_copy = df_netflix_copy.explode('listed_in')

# Remove spaces before and after the countries
df_netflix_copy['listed_in'] = df_netflix_copy['listed_in'].str.strip()

# Get the unique values from the 'country' column
unique_values_listed = df_netflix_copy['listed_in'].unique()

# Print the unique values
print(unique_values_listed)
```



Fact : Using .copy() before applying .explode() ensures that the original DataFrame remains unchanged, allowing for safe manipulation without altering the source data.

##### Wordcloud

A word cloud is a visual representation of text data where the size of each word indicates its frequency or importance within the text. 

In a word cloud:

- Words from the text are displayed in varying sizes, with more frequently occurring words typically shown in larger font sizes.
- The layout of the words is usually random or arranged in a visually appealing way, often with more important words placed prominently.
- Word clouds are commonly used to quickly grasp the most significant terms or themes present in a body of text, providing a visual summary of the textual data.


They are popular in data visualization for presentations, reports, and online content, as they offer an intuitive way to visualize textual information and highlight key insights or trends.


##### Title column

```python
nltk.download('punkt')
import string
from nltk.corpus import stopwords as nltk_stopwords_en
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
# Tokenize the text in the 'title' column
df_netflix['title_tokens'] = df_netflix['title'].apply(lambda x: word_tokenize(x.lower()))

print(df_netflix[['title', 'title_tokens']].head())
```

```python
# Filter out punctuation from each list of tokens
df_netflix['title_tokens_nopunctuation'] = df_netflix['title_tokens'].apply(lambda tokens: [word for word in tokens if word not in string.punctuation])

# Join the tokens into a single text
df_netflix['texte_propre'] = df_netflix['title_tokens_nopunctuation'].apply(lambda tokens: ' '.join(tokens))

# Tokenize the clean text
df_netflix['texte_propre_split'] = df_netflix['texte_propre'].apply(lambda x: nltk.FreqDist(x.split(" ")))

##### StopWords
# Define a list of French stopwords
stopwords_fr = nltk.corpus.stopwords.words("french")
# Define a list of English stopwords
stopwords_en = nltk_stopwords_en.words("english")
# Concatenate French and English stopwords
stopwords_all = stopwords_fr + stopwords_en
# Remove stopwords
df_netflix['tokens_clean'] = df_netflix['texte_propre_split'].apply(lambda freq_dist: [word for word in freq_dist.keys() if word not in stopwords_all])

# Remove "'s"
df_netflix['tokens_clean'] = df_netflix['tokens_clean'].apply(lambda tokens: [word for word in tokens if word != "'s"])

#####Frequency 
# Calculate frequency distribution for clean tokens
df_netflix['freq_clean'] = df_netflix['tokens_clean'].apply(lambda tokens: nltk.FreqDist(tokens))

# Display the DataFrame with cleaned tokens and frequency distributions
print(df_netflix[['title', 'title_tokens', 'title_tokens_nopunctuation', 'texte_propre', 'texte_propre_split', 'tokens_clean', 'freq_clean']].head())
```

```python
# Convert frequency distribution to dictionary
word_freq_text = ' '.join([' '.join([word] * freq) for freq_dist in df_netflix['freq_clean'] for word, freq in freq_dist.items()])

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', colormap='viridis').generate(word_freq_text)

plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.margins(x=0, y=0)
plt.show()
```

This WordCloud shows the frequency of each word included in the title. 

```python
# Get the top 20 words and their frequencies
top_words = df_netflix['freq_clean'].apply(lambda freq_dist: dict(freq_dist.most_common(20)))

# Combine the dictionaries into one
combined_dict = {}
for word_freq in top_words:
    for word, freq in word_freq.items():
        combined_dict[word] = combined_dict.get(word, 0) + freq

# Sort the combined dictionary by frequency
sorted_dict = dict(sorted(combined_dict.items(), key=lambda item: item[1], reverse=True))

# Take only the top 20 words
top_20_words = dict(list(sorted_dict.items())[:20])

# Plot the top 20 words
plt.figure(figsize=(10, 6))
plt.bar(top_20_words.keys(), top_20_words.values())
plt.title('Top 20 Most Frequent Words in Title')
plt.xlabel('Word')
plt.ylabel('Frequency')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

##### Heatmap for Correlation Analysis

A heatmap correlation visually represents the correlation matrix of a dataset using colors.

Each cell in the heatmap represents the correlation coefficient between two variables.

The correlation coefficient measures the strength and direction of the linear relationship between two variables. It ranges from -1 to 1.

- A value close to 1 indicates a strong positive correlation (as one variable increases, the other tends to increase).
- A value close to -1 indicates a strong negative correlation (as one variable increases, the other tends to decrease).
- A value close to 0 indicates a weak correlation or no correlation between the variables.

The colors in the heatmap indicate the magnitude and direction of the correlation:
- Shades of red represent positive correlations.
- Shades of blue represent negative correlations.
- Lighter shades indicate stronger correlations, while darker shades indicate weaker correlations.

```python
numeric_columns = df_netflix.select_dtypes(include=['int64', 'float64'])
non_numeric_columns = df_netflix[['type']]

# Perform one-hot encoding
encoded_df = pd.get_dummies(non_numeric_columns)

# Concatenate encoded DataFrame with numeric columns
final_df = pd.concat([encoded_df, numeric_columns], axis=1)

# Calculate the correlation matrix
correlation_matrix = final_df.corr()

# Create a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Heatmap of Netflix Dataset')
plt.show()
```

Variables:
The heatmap considers 3 variables:
Type: Whether the content is a movie or a TV show.
Release Year: The year when the content was released.
Duration: The length of the content (in minutes).

Observations : 
- Movies have a strong positive correlation with duration. Longer movies are common.
- TV shows exhibit a strong negative correlation with duration. TV shows are typically shorter.

Others correlation are too weak to be explained.
